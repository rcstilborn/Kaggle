```python
# Delete the data we don't need
ignore_cols = ["PassengerId", "Name", "Ticket", "Embarked", "Cabin"]
train_data = train_data.drop(ignore_cols, axis=1)
test_data = test_data.drop(ignore_cols, axis=1)
train_data.head()
```

```python
 from sklearn.preprocessing import Imputer
    
def nan_replacement(data, columns):
    for column in columns:
        imp = Imputer()
        data[column] = imp.fit_transform(data[column].values.reshape(-1,1))
    return data

nan_cols = ["Age","SibSp","Parch"]
train_data = nan_replacement(train_data,nan_cols)
test_data = nan_replacement(test_data,nan_cols)
```
